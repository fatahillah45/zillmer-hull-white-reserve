```python
df_tmi = pd.read_csv("TMI IV (2019) - L .csv")
df_tmi
```

```python
df_tmi['lx'] = df_tmi['lx'].str.replace(',', '.').astype(float)
df_tmi['dx'] = df_tmi['dx'].str.replace(',', '.').astype(float)
lx = df_tmi['lx'].values
dx = df_tmi['dx'].values
```

```python
def anuitas_berjangka_awal(x, m, v, lx):
    anuitas = 0.0
    lx_awal = lx[x]
    for k in range(0, m):
        v_k = v[k]
        k_p_x = lx[x + k] / lx_awal
        anuitas += v_k * k_p_x
    return anuitas
```

```python
def premi_tunggal_bersih(x, n, v, lx, dx):
    l_x_awal = lx[x]
    sum_numerator = 0.0
    for t in range(1, n + 1):
        usia_kematian_awal = x + t - 1
        nu_t = v[t]
        d_x_t_1 = dx[usia_kematian_awal]
        sum_numerator += nu_t * d_x_t_1
    A_x_n_1 = sum_numerator / l_x_awal

    return A_x_n_1
```

```python
def f_zillmer(x,n,m, v,lx,dx):
  premi_tahunan_zillmer = premi_tunggal_bersih(x+1,n-1,v,lx,dx) / anuitas_berjangka_awal(x+1,m-1,v,lx)
  C_x = dx[x] * v[1]
  D_x = lx[x] * 1
  f = premi_tahunan_zillmer - (C_x/D_x)
  return f
```

```python
def cadangan_zillmer_list(x, n, m, v, lx, dx, f):
    Vz_list = []
    P = premi_tunggal_bersih(x, n, v, lx, dx) / anuitas_berjangka_awal(x, m, v, lx)
    a_awal = anuitas_berjangka_awal(x, m, v, lx)
    P_zillmer = P + (f / a_awal)
    for t in range(1, n + 1):
        A_t = premi_tunggal_bersih(x + t, n - t, v, lx, dx)
        sisa_m = m - t
        if sisa_m <= 0:
            a_t = 0.0
        else:
            a_t = anuitas_berjangka_awal(x + t, sisa_m, v, lx)
        Vz = A_t - P_zillmer * a_t
        Vz_list.append(Vz)
    return Vz_list
```

```python
nama_file_input = 'Lampiran_1.csv'
df_simulasi = pd.read_csv(nama_file_input)

df_simulasi['Santunan'] = df_simulasi['Santunan'].str.replace('Rp', '', regex=False).str.replace('.', '', regex=False).str.replace(',', '.', regex=False).astype(float)

for index, row in df_simulasi.iterrows():
    x = int(row['Usia'])
    n = int(row['Masa Pertanggungan'])
    m = int(row['Masa Bayar'])
    UP = row['Santunan']

    anuitas_awal = anuitas_berjangka_awal(x, m, v, lx)
    A_x_n = premi_tunggal_bersih(x, n, v, lx, dx)
    premi_tunggal_rp = A_x_n * UP
    premi_tahunan_rp = (A_x_n / anuitas_awal) * UP

    f_val = f_zillmer(x, n, m, v, lx, dx)

    cadangan_per_unit = cadangan_zillmer_list(x, n, m, v, lx, dx, f_val)
    cadangan_rupiah = [val * UP for val in cadangan_per_unit]
    df_simulasi.at[index, 'Anuitas'] = anuitas_awal
    df_simulasi.at[index, 'Premi_Tunggal'] = premi_tunggal_rp
    df_simulasi.at[index, 'Premi_Tahunan'] = premi_tahunan_rp
    df_simulasi.at[index, 'f_Zillmer'] = f_val
    df_simulasi.at[index, 'V_t1'] = max(0, cadangan_rupiah[0]) if len(cadangan_rupiah) >= 1 else 0
    df_simulasi.at[index, 'V_t5'] = max(0, cadangan_rupiah[4]) if len(cadangan_rupiah) >= 5 else 0
    df_simulasi.at[index, 'V_Max'] = max(0, max(cadangan_rupiah))
df = df_simulasi.copy()
def format_rupiah(angka):
    if angka < 0:
        return f"-Rp {abs(angka):,.2f}".replace(',', 'x').replace('.', ',').replace('x', '.')
    else:
        return f"Rp {angka:,.2f}".replace(',', 'x').replace('.', ',').replace('x', '.')

kolom_uang = ['Santunan', 'Premi_Tunggal', 'Premi_Tahunan', 'V_t1', 'V_t5', 'V_Max']
for col in kolom_uang:
    df[col + '_Fmt'] = df[col].apply(format_rupiah)

df['Anuitas_Fmt'] = df['Anuitas'].apply(lambda val: str(val).replace('.', ','))
df['f_Zillmer_Fmt'] = df['f_Zillmer'].apply(lambda val: str(val).replace('.', ','))

kolom_dasar = ['Kode Polis', 'Usia', 'Masa Pertanggungan', 'Masa Bayar', 'Santunan_Fmt']
kolom_header_dasar = ['Kode Polis', 'Usia (x)', 'Masa Pertanggungan (n)', 'Masa Bayar (m)', 'Santunan (S)']

tabel_anuitas = df[kolom_dasar + ['Anuitas_Fmt']].copy()
tabel_anuitas.columns = kolom_header_dasar + ['Anuitas Berjangka']

tabel_premi_tunggal = df[kolom_dasar + ['Premi_Tunggal_Fmt']].copy()
tabel_premi_tunggal.columns = kolom_header_dasar + ['Premi Tunggal Bersih']

tabel_premi_tahunan = df[kolom_dasar + ['Premi_Tahunan_Fmt']].copy()
tabel_premi_tahunan.columns = kolom_header_dasar + ['Premi Tahunan Bersih']

tabel_cadangan = df[kolom_dasar + ['V_t1_Fmt', 'V_t5_Fmt', 'V_Max_Fmt']].copy()
tabel_cadangan.columns = kolom_header_dasar + ['Cadangan Akhir Th-1', 'Cadangan Akhir Th-5', 'Cadangan Maksimum']

tabel_f_zillmer = df[kolom_dasar + ['f_Zillmer_Fmt']].copy()
tabel_f_zillmer.columns = kolom_header_dasar + ['Nilai f Zillmer']


nama_file_excel = 'Tabel_Skripsi_Bab_4_Lengkap_V2.xlsx'
with pd.ExcelWriter(nama_file_excel) as writer:
    tabel_anuitas.to_excel(writer, sheet_name='Anuitas', index=False)
    tabel_premi_tunggal.to_excel(writer, sheet_name='Premi Tunggal', index=False)
    tabel_premi_tahunan.to_excel(writer, sheet_name='Premi Tahunan', index=False)
    tabel_cadangan.to_excel(writer, sheet_name='Cadangan Zillmer', index=False)
    tabel_f_zillmer.to_excel(writer, sheet_name='f Zillmer', index=False)
```

```python
data_tabel_cadangan = {
    't (Tahun)': list(range(1, 21))
}

for index, row in df_simulasi.iterrows():

    x = int(row['Usia'])
    n = int(row['Masa Pertanggungan'])
    m = int(row['Masa Bayar'])
    UP = float(row['Santunan'])

    f_val = f_zillmer(x, n, m, v, lx, dx)
    cadangan_per_unit = cadangan_zillmer_list(x, n, m, v, lx, dx, f_val)

    cadangan_rupiah = [max(0, val * UP) for val in cadangan_per_unit]

    kolom_polis_ini = []

    for t_index in range(20):
        if t_index < n:
            nilai = cadangan_rupiah[t_index]
            nilai_rp = f"Rp {nilai:,.2f}".replace(',', 'x').replace('.', ',').replace('x', '.')
            kolom_polis_ini.append(nilai_rp)
        else:
            kolom_polis_ini.append("-")

    kode_polis = f"P-{str(index+1).zfill(3)}"
    data_tabel_cadangan[kode_polis] = kolom_polis_ini

df_tabel_cadangan_final = pd.DataFrame(data_tabel_cadangan)

nama_file = 'Tabel_Cadangan_Format_Kolom_Polis.xlsx'
df_tabel_cadangan_final.to_excel(nama_file, index=False)
print(df_tabel_cadangan_final.iloc[:, :6].head())
```
